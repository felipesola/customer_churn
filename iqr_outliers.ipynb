```python
# How many rows would remain in each feature after removing the outliers using IQR method
for column in df.columns:

    Q1 = df[column].quantile(.25)
    Q3 = df[column].quantile(.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_without_outliers = df[(df[column] < upper) & (df[column] > lower)]
    df_shape = [column, df_without_outliers.shape]

    print(df_shape)
```

```python
# Remove the outliers using IQR method of a selected array of features
selected_columns = df[['Call Failure', 'Subscription Length', 'Charge Amount', 'Seconds of Use', 'Frequency of use',\
              'Frequency of SMS', 'Distinct Called Numbers', 'Age Group', 'Age', 'Customer Value']]

#df_to_iterate = df
for column in selected_columns.columns:
    Q1 = selected_columns[column].quantile(.25)
    Q3 = selected_columns[column].quantile(.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(selected_columns[column] < upper) & (selected_columns[column] > lower)]

df.shape
```

```python
# Save a new data set without the outliers in the 'datasets' folder in order to run the model and evaluate it
df.to_csv('.\datasets\customer_churn_iqr_removed.csv', index = False)
```
